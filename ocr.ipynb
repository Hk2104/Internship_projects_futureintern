```python
from PIL import Image
import pytesseract

def extract_text(image_path):
    image = Image.open(image_path)
    text = pytesseract.image_to_string(image, lang='eng+hin')  # Extract text in both languages
    return text
```

```python
import gradio as gr

def ocr_function(image):
    text = extract_text(image.name)
    return text

def search_in_text(text, keyword):
    if keyword in text:
        return f"Keyword '{keyword}' found in the text!"
    else:
        return f"Keyword '{keyword}' not found."

with gr.Blocks() as app:
    gr.Markdown("### OCR Image Upload and Search")
    image_input = gr.Image(type="filepath")
    text_output = gr.Textbox(label="Extracted Text")
    keyword_input = gr.Textbox(label="Enter Keyword to Search")
    search_output = gr.Textbox(label="Search Result")

    image_input.change(ocr_function, inputs=image_input, outputs=text_output)
    keyword_input.change(search_in_text, inputs=[text_output, keyword_input], outputs=search_output)

app.launch()
```
